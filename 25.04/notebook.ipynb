читаем список песен и парсим тексты с genius

```python
import csv
import requests
from bs4 import BeautifulSoup
```

```python
songs = []
with open('spotify_data.csv', encoding='utf-8') as f:
    for row in csv.DictReader(f):
        songs.append((row['artist_name'], row['track_name']))
songs[:10]
```

```python
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0 Safari/537.36'
}

URL = 'https://genius.com/Dj-paul-omg-lyrics'
res = requests.get(URL, headers=HEADERS)
soup = BeautifulSoup(res.text, 'html.parser')
soup.find('h1').text
```

```python
def fetch_lyrics_page(url):
    res = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(res.text, 'html.parser')
    blocks = soup.select('div[data-lyrics-container="true"]')
    return '\n'.join(b.get_text('\n') for b in blocks)
```
