открыть файл, отсортировать по числу уникальных треков и вывести топ

```python
from csv import DictReader

art_songs = {}
with open('dataset.csv', encoding='utf-8') as f:
    for row in DictReader(f):
        for artist in row['artists'].split(';'):
            artist = artist.strip()
            if artist in art_songs:
                art_songs[artist].add(row['track_name'])
            else:
                art_songs[artist] = {row['track_name']}

art_songs_sorted = sorted(art_songs.items(), key=lambda x: len(x[1]), reverse=True)
for artist, tracks in art_songs_sorted[:10]:
    print(artist, len(tracks))
```

суммарная популярность по исполнителям: топ-50 и аутсайдеры

```python
pop = {}
with open('dataset.csv', encoding='utf-8') as f:
    for row in DictReader(f):
        popularity = int(row['popularity'])
        for artist in row['artists'].split(';'):
            artist = artist.strip()
            pop[artist] = pop.get(artist, 0) + popularity

most = sorted(pop.items(), key=lambda x: x[1], reverse=True)[:50]
least = sorted(pop.items(), key=lambda x: x[1])[:10]
most[:10]
```

число хитов (популярность > 67) у каждого и самые непопулярные по макс. популярности

```python
art_pop = {}
with open('dataset.csv', encoding='utf-8') as f:
    for row in DictReader(f):
        popularity = int(row['popularity'])
        for artist in row['artists'].split(';'):
            artist = artist.strip()
            art_pop.setdefault(artist, []).append(popularity)

hits = {a: sum(1 for p in ps if p > 67) for a, ps in art_pop.items()}
max_pop = {a: max(ps) for a, ps in art_pop.items()}

print('hit makers')
for artist, n in sorted(hits.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(artist, n)
print('\nlosers')
for artist, p in sorted(max_pop.items(), key=lambda x: x[1])[:10]:
    print(artist, p)
```

с кем больше всего фитов: по совместным трекам (несколько исполнителей в поле artists)

```python
art_fits = {}
with open('dataset.csv', encoding='utf-8') as f:
    for row in DictReader(f):
        artists = [a.strip() for a in row['artists'].split(';')]
        if len(artists) >= 2:
            for artist in artists:
                art_fits[artist] = art_fits.get(artist, 0) + 1

sorted(art_fits.items(), key=lambda x: x[1], reverse=True)[:10]
```

и по упоминаниям feat./ft. в названии трека

```python
import csv

names = []
with open('dataset.csv', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    track_names = [row['track_name'] for row in reader]

feated = []
for name in track_names:
    low = name.lower()
    for marker in ('feat. ', 'ft. '):
        if marker in low:
            tail = name[low.index(marker) + len(marker):]
            for stop in (')', ']'):
                if stop in tail:
                    tail = tail[:tail.index(stop)]
            for part in tail.split(','):
                if part.strip():
                    feated.append(part.strip())

feat_count = {}
for a in feated:
    feat_count[a] = feat_count.get(a, 0) + 1
sorted(feat_count.items(), key=lambda x: x[1], reverse=True)[:10]
```
