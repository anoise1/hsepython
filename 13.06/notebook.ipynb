```python
import pandas as pd
import numpy as np
import random
from statsmodels.stats.weightstats import ztest
```

```python
data = pd.read_csv('train_data.csv')
data
```

```python
data_p1 = data[data['Pclass_1'] == 1]
data_p23 = data[(data['Pclass_2'] == 1) | (data['Pclass_3'] == 1)]
data_p1.describe()
```

```python
data_p23.describe()
```

```python
data_p1['Survived'].mean(), data_p23['Survived'].mean()
```

бутстрэп доверительного интервала разницы долей выживших

```python
ships = []

for _ in range(10000):
    p1 = pd.Series(random.choices(data_p1['Survived'].values, k=data_p1.shape[0])).mean()
    p23 = pd.Series(random.choices(data_p23['Survived'].values, k=data_p23.shape[0])).mean()
    ships.append(p1 - p23)
```

```python
np.quantile(ships, q=[0.025, 0.975])
```

ручная z-статистика и z-тест

```python
(data_p1['Survived'].mean() - data_p23['Survived'].mean()) / ((data['Survived'].mean() * (1 - data['Survived'].mean())) / (len(data['Survived']))) ** 0.5
```

```python
ztest(data_p1.Survived, data_p23.Survived)
```
